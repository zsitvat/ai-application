```python
from openai import OpenAI
client = OpenAI()

fine_tune_job_id = "fine_tuning_job_id"

# List 10 fine-tuning jobs
client.fine_tuning.jobs.list(limit=10)

# Retrieve the state of a fine-tune
client.fine_tuning.jobs.retrieve(fine_tune_job_id)

# Cancel a job
client.fine_tuning.jobs.cancel(fine_tune_job_id)

# List up to 10 events from a fine-tuning job
client.fine_tuning.jobs.list_events(fine_tuning_job_id=fine_tune_job_id, limit=10)

# Delete a fine-tuned model
client.models.delete("fine_tuned_model_id")
```
