```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import random
import json
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

#https://www.geeksforgeeks.org/intent-recognition-using-tensorflow/
```

```python

file_path = 'intents.xlsx'

sheet_name = 'Tanító mondatok'
df = pd.read_excel(file_path, sheet_name=sheet_name)

intent_dict = { }

# Collect the data
for _, row in df.iterrows():
    intent = row['Intent']
    question = row['Kérdések']
    if intent not in intent_dict:
        intent_dict[intent] = []
    intent_dict[intent].append(question)

# Display the collected data
print("Intents: ",intent_dict)
print("Number of intents: ", len(intent_dict))
```

```python
def clean(line):
    cleaned_line = ''
    for char in line:
        if char.isalpha():
            cleaned_line += char
        else:
            cleaned_line += ' '
    cleaned_line = ' '.join(cleaned_line.split())
    return cleaned_line.lower() # Kisbetűsítés
```

```python
#list of intents
intents = []                                            
unique_intents = []
#all text data to create a corpus
text_input= []    
                            
for intent,questions in intent_dict.items():
    #list of unique intents
    if intent not in unique_intents:            
        unique_intents.append(intent)  
    for question in questions:
        #cleaning is done before adding text to corpus
        text_input.append(clean(question))                    
        intents.append(intent)

```

```python
print("Intent :",intents[0])
print("Number of Intent:",len(intents))
print("Sample Input:", text_input[0])
print('Length of text_input:',len(text_input))
```

```python
tokenizer = Tokenizer(filters='',oov_token='<unk>')
tokenizer.fit_on_texts(text_input)
sequences = tokenizer.texts_to_sequences(text_input)
#padded_sequences = pad_sequences(sequences, padding='pre')
padded_sequences = pad_sequences(sequences, padding='post')
print('Shape of Input Sequence:',padded_sequences.shape)
padded_sequences[:5]
```

```python
intent_to_index = {}
categorical_target = []
index = 0

for intent in intents:
    if intent not in intent_to_index:
        intent_to_index[intent] = index
        index += 1
    categorical_target.append(intent_to_index[intent])

num_classes = len(intent_to_index)
print('Number of Intents :',num_classes)

# Convert intent_to_index to index_to_intent
index_to_intent = {index: intent for intent, index in intent_to_index.items()}
index_to_intent
```

```python
categorical_vec = tf.keras.utils.to_categorical(categorical_target, 
                                                num_classes=num_classes)

categorical_vec = categorical_vec.astype('int32')

print('Shape of Ca',categorical_vec.shape)
categorical_vec[:5]
```

```python
epochs=100
embed_dim=300
lstm_num=50
output_dim=categorical_vec.shape[1]
input_dim=len(unique_intents)
print("Input Dimension :{},\nOutput Dimension :{}".format(input_dim,output_dim))
```

```python
#model = tf.keras.models.Sequential([
#    tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, embed_dim),
#    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_num, dropout=0.1)),  
#    tf.keras.layers.Dense(lstm_num, activation='relu'),
#    tf.keras.layers.Dropout(0.4),
#    tf.keras.layers.Dense(output_dim, activation='softmax')
#])

model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(len(tokenizer.word_index) + 1, 300),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True, dropout=0.2)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, dropout=0.2)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(len(unique_intents), activation='softmax')
])

#optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
#optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.001)
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.01)
optimizer = tf.keras.optimizers.Adadelta(learning_rate=1.0)
optimizer = tf.keras.optimizers.Adamax(learning_rate=0.002)
optimizer = tf.keras.optimizers.Nadam(learning_rate=0.002)
optimizer = tf.keras.optimizers.Ftrl(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(padded_sequences, categorical_vec, epochs=epochs, verbose=0)
```

```python



sample_df = df.sample(n=100, random_state=42)

test_text_inputs = sample_df["Kérdések"].tolist()
test_intents = sample_df['Intent'].tolist()


# Print the extracted data
print("Test Text Inputs:", test_text_inputs)
print("Test Intents:", test_intents)

test_sequences = tokenizer.texts_to_sequences(test_text_inputs)
test_padded_sequences = pad_sequences(test_sequences,  padding='pre')
test_labels = np.array([unique_intents.index(intent) for intent in test_intents])
test_labels = tf.keras.utils.to_categorical(test_labels, num_classes=num_classes)
loss, accuracy = model.evaluate(test_padded_sequences, test_labels)
```

```python
def response(sentence):
    sent_tokens = []
    # Split the input sentence into words
    words = sentence.split()
    # Convert words to their corresponding word indices
    for word in words:                                           
        if word in tokenizer.word_index:
            sent_tokens.append(tokenizer.word_index[word])
        else:
            # Handle unknown words
            sent_tokens.append(tokenizer.word_index['<unk>'])
    sent_tokens = tf.expand_dims(sent_tokens, 0)
    #predict numerical category
    pred = model(sent_tokens)    
    #category to intent
    pred_class = np.argmax(pred.numpy(), axis=1)     
    
    print("Predicted Class:", index_to_intent[pred_class[0]])           
    
    return random.choice(
        intent_dict[index_to_intent[pred_class[0]]]), index_to_intent[pred_class[0]]
```

```python
def response(sentence):
    sent_tokens = tokenizer.texts_to_sequences([clean(sentence)])
    sent_tokens = pad_sequences(sent_tokens, padding='post', maxlen=padded_sequences.shape[1])
    pred = model.predict(sent_tokens)
    pred_class = np.argmax(pred, axis=1)[0]
    return random.choice(intent_dict[unique_intents[pred_class]]), unique_intents[pred_class]
```

```python
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate

from dotenv import load_dotenv
import os

load_dotenv(".env")

def run_ai(user_input):

    llm =  AzureChatOpenAI(
                azure_endpoint=os.environ.get("AZURE_BASE_URL"),
                openai_api_version=os.environ.get("AZURE_API_VERSION"),
                azure_deployment="chatboss_sweden-central_chat_gtp4o_mini",
                temperature=0
            )



    prompt = ChatPromptTemplate.from_messages([
        ("system", "Rephrase the question"),
        ("user", "{user_input}")
    ])
    
    chain = llm | prompt
    
    return chain.invoke(input=user_input)

run_ai(user_input="What is the weather today?")
    
```

```python
random_questions = df['Kérdések'].sample(n=10, random_state=42).tolist()

correct_intents = 0

for _, row in random_questions.iterrows():
    original_question = row['Kérdések']
    original_intent = row['Intent']
    
    query = run_ai(original_question)
    print(f"Original Question: {original_question}")
    print(f"Rephrased Question: {query}")
    print()
    
    bot_response, predicted_intent = response(query)
    print('Random intent question: {}  \nIntent: {}'.format(bot_response, predicted_intent))
    print()
    
    if predicted_intent == original_intent:
        correct_intents += 1

total_questions = len(random_questions)
accuracy = (correct_intents / total_questions) * 100

print(f"Total Questions: {total_questions}")
print(f"Correct Intents: {correct_intents}")
print(f"Accuracy: {accuracy:.2f}%")
```
