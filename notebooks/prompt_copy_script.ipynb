```python
from langfuse import Langfuse
from dotenv import load_dotenv,find_dotenv

load_dotenv(find_dotenv())

def get_prompt(prompt_name):
    # Initialize Langfuse client
    langfuse = Langfuse()
 
    # Get current `production` version of a text prompt
    prompt = langfuse.get_prompt(prompt_name, label="latest")

    return prompt
```

```python
def upload_prompt(prompt_name, prompt,prompt_type = "chat"):
    # Initialize Langfuse client
    langfuse = Langfuse(
        host="https://langfuse.attrecto-dev.com",
        public_key="",
        secret_key="",
        
                        )
    
    print(f"Uploading prompt: {prompt_name}")
    print(f"Type: {prompt_type}")
    
    errored_prompts = []

    if prompt_type=="text":
        prompt=str(prompt)
 
    try:
        langfuse.create_prompt(name=prompt_name.replace("-gpt-35-0125","").replace("-0125",""), prompt=prompt, type=prompt_type)

    except Exception as e:
        print(f"Error uploading prompt: {e}")
        errored_prompts.append(prompt_name)

    return errored_prompts
```

```python
for prompt_name in prompts:
    if "trocellen" in prompt_name:
        prompt = get_prompt(prompt_name)

        prompt_type = "text" if "router" in prompt_name or "summarizer" in prompt_name or "standalone" in prompt_name or prompt_name=="translate" else "chat"

        errored_promps = upload_prompt(prompt_name, prompt.prompt,prompt_type=prompt_type)

if errored_promps:
    print(f"Error uploading prompts: {errored_promps}")
else:

    print("All prompts uploaded successfully")
```
